```python
data.columns = ['Latitude', 'Longitude', '0', 'Altitude', 'NumDays', 'Date', 'Time']
data
```

```python
data["Timestamp"] = data["Date"].map(str) + " " + data["Time"]
data.head()
```

```python
data.Timestamp = pd.to_datetime(data.Timestamp)
data.head()
data.dtypes
```

```python
import numpy as np
import matplotlib.pyplot as plt

data.plot(x='Timestamp', y='Latitude', style='y^-',  linestyle=':', figsize=(20,10))
plt.title('Trend - User 000')
plt.ylabel('Date')
plt.xlabel('Time-Hour')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

color_dict = {'Latitude': '#FF0000', 'Longitude': '#0000FF'}
data.plot(x='Latitude', y='Longitude', color=[color_dict.get(x, '#333333') for x in data.columns], kind = 'scatter', figsize=(20,10))
plt.title('Trend - User 000')
plt.ylabel('Longitude')
plt.xlabel('Latitude')
plt.show()
```
