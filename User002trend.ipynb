```python
import glob

filenames = glob.glob("/home/shashank/Downloads/Geolife Trajectories 1.3/Data/002/Trajectory/200*.plt")
filenames
```

```python
import pandas as pd

list_of_dfs = [pd.read_csv(filename, skiprows=6, header = None) for filename in filenames]


list_of_dfs
```

```python
import pandas as pd
import numpy as np


combined_df = pd.concat(list_of_dfs, ignore_index=True)
combined_df.columns = ['Latitude', 'Longitude', '0', 'Altitude', 'NumDays', 'Date', 'Time']
combined_df
```

```python
for filename in filenames:
    combined_df['Filename'] = filename[-18:]
combined_df.Latitude.size
```

```python
combined_df["Timestamp"] = combined_df["Date"].map(str) + " " + combined_df["Time"]
combined_df.Timestamp = pd.to_datetime(combined_df.Timestamp)
combined_df.dtypes
```

```python

combined_df.index = pd.to_datetime(combined_df.Timestamp)
```

```python
df1 = combined_df.resample('H').last()
df1
```

```python
import numpy as np
import matplotlib.pyplot as plt

combined_df.plot(x='Timestamp', y= '0' , style='o', figsize=(20,10))
plt.title('Trend - User 000')
plt.xlabel('Timestamp')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

combined_df.plot(x='Timestamp', y= 'Latitude' , style='o', figsize=(20,10))
plt.title('Trend - User 000')
plt.ylabel('Latitude')
plt.xlabel('Timestamp')
plt.show()
```

```python
df1['Weekday'] = df1['Timestamp'].dt.dayofweek
df1.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

df1.plot(x='Timestamp', y= 'Weekday' , style='o', figsize=(20,10), markersize=15, legend=True, colormap='Spectral')
plt.title('Trend - User 000')
plt.ylabel('Weekday')
plt.xlabel('Timestamp')
plt.show()
```

```python
df1['Day of week'] = df1['Timestamp'].dt.weekday_name
df1['Day of week'].value_counts()
```

```python
df1['DateInt'] = df1['Timestamp'].map(lambda x: (100*x.year + x.month)*100 + x.day)
df1['HourInt'] = df1['Timestamp'].dt.hour
df1.plot(x='HourInt', y= 'DateInt' , style='^', figsize=(20,20), markersize=10, legend=True, colormap='Spectral')
plt.title('Trend - User 000')
plt.ylabel('Dates')
plt.xlabel('Time')
plt.show()
```
